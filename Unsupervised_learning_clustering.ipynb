```python
import pandas as pd
import numpy as np

data=pd.read_csv('rain-dataset.csv').values
print(data,len(data))
```

```python
x=data[:,0]
y=data[:,1]

from matplotlib import pyplot as plt
plt.scatter(x,y)
plt.xlabel('Temp')
plt.ylabel('Humanity')
plt.title('Scatter Plot')
plt.show()
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=2)
model.fit(data)
labels=model.labels_
print('clustered numbers :',labels,len(labels))
```

```python
from matplotlib import pyplot as plt
for i in range(len(data)):
  if(labels[i]==0):
    plt.scatter(data[i,0],data[i,1],color='blue')
  elif(labels[i]==1):
    plt.scatter(data[i,0],data[i,1],color='red')
```

```python
model=KMeans(n_clusters=3)
model.fit(data)
labels=model.labels_

print('dataset :',data)
print('Clusterred labels:',labels)
```

```python
for i in range(len(data)):
  if(labels[i]==0):
    plt.scatter(data[i,0],data[i,1],color='blue')
  elif(labels[i]==1):
    plt.scatter(data[i,0],data[i,1],color='red')
  elif(labels[i]==2):
    plt.scatter(data[i,0],data[i,1],color='green')
```

```python
cc1,cc2,cc3=model.cluster_centers_
print('cc1:',cc1)
print('cc2:',cc2)
print('cc3:',cc3)
```

```python
for i in range(len(data)):
   if(labels[i]==0):
         plt.scatter(data[i,0],data[i,1],c='b')
   else:
         plt.scatter(data[i,0],data[i,1],c='r')

plt.scatter(cc1[0],cc1[1],marker='x',c='b')
plt.scatter(cc2[0],cc2[1],marker='x',c='r')
plt.scatter(cc3[0],cc3[1],marker='x',c='g')
plt.show()
```

```python
plt.scatter(data[:,0],data[:,1])
plt.xlabel('Temp')
plt.ylabel('Humanity')
plt.savefig('plot.png',dpi=300)
plt.show()
```

```python
import pandas as pd
dataset=pd.read_csv('iris.csv').values

print(dataset)
```

```python
data=dataset[:,0:4]
print(data)
```

```python
from sklearn.cluster import MeanShift
model=MeanShift(bandwidth=0.85)
model.fit(data)
labels=model.labels_

centers=model.cluster_centers_

print('shape of cluster center:',centers.shape)
print('cluster centers:',centers)
```

```python
label=model.labels_
print('0-50',labels[:50])
print('50-100',labels[50:100])
print('100-150',labels[100:150])
```
