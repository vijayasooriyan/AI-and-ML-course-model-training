```python
from keras.models import load_model
model=load_model('/content/drive/MyDrive/FFNN-iris.keras')
model.summary()
```

```python
import numpy as np
test_data=np.array([[5.0,3.2,1.4,0.1]])
result=model.predict(test_data)
print(result)
```

```python
label=np.argmax(result,axis=1 )
print(label)
```

```python
prob=np.max(result,axis=1)
print(prob)
```
